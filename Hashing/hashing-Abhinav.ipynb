### Collision Resolution Method: Opening Addressing (Linear Probing)

```python
class HashTable:
    
    def __init__(self, m):
        self.m = m
        self.hashtable = self.create_hash_table()
    
    def create_hash_table(self):
        return [ None for _ in range(self.m) ]
    
    def _prehash(self, key):
        #challenge: handle negative keys and string
        if (type(key) == str):
            key = hash(key)  #returns a number for you
            
        if ((type(key) == int) | (type(key) == float)):
            if (key < 0):
                key = hash(float(key)) * -1  #first convert to float, then hash it
        
        assert (key > 0) & (type(key) == int)
    
        return key
    
    def _hash(self, key):
        #get the position using division method
        index = key % self.m
        return index
    
    def _search_indices_comparision(self, start, end, found, position, value, mode, key):
        for i in range(start, end):
            if (self.hashtable[i]) == None or (self.hashtable[i] == -9999 and mode == "insert"): # -9999 --> delete flag
                position = i
                break
            elif self.hashtable[i] == -9999:
                continue
            elif self.hashtable[i][0] == key:
                found = True 
                position = i
                value = self.hashtable[i]
                break
        
        return found, position, value
    
    def search(self, key, mode):
        key = self._prehash(key)
        index = self._hash(key)
        
        value    = None
        position = None
        found    = False
        
        found, position, value = self._search_indices_comparision(index, len(self.hashtable), found, position, value, mode, key)
        
        if not found and position is None:
            found, position, value = self._search_indices_comparision(0, index, found, position, value, mode, key)
        
        return found, position, value
    
    def insert(self, key, val):
        found, position, return_val = self.search(key, "insert")
        if found:
            self.hashtable[position] = (key , val)
        else:
            self.hashtable[position] = (key , val)
        
        print("--- INSERT OPERATION --")
        print(self.hashtable, end="\n\n")
    
    def delete(self, key):
        found, position, _ = self.search(key, "delete")
        
        if found:
            self.hashtable[position] = -9999
        else:
            print("Key is not found")
        
        print("--- DELETE OPERATION --")
        print(self.hashtable, end="\n\n")
```
