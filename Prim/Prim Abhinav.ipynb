# Prim's Algorithm with Adjacency List

```python
G = {
    0: [[1, 2], [2, 2], [3, 5], [4, 3]],
    1: [[0, 2], [2, 1], [3, 4], [4, 4]],
    2: [[0, 2], [1, 1], [3, 3], [4, 5]],
    3: [[0, 5], [1, 4], [2, 3], [4, 9]],
    4: [[0, 3], [1, 4], [2, 5], [3, 9]]
}

INF = 9999
N   = 5
r   = 0

from heapdict import heapdict

Q = heapdict()
for i in range(N):
    Q[i] = INF
Q[r] = 0

print(f"Q:  {list(Q)}")

pi = [None] * 5
pi[r] = -1
print(f"PI: {pi}")

def v_in_Q(Q, v):
    #get the keys in Q
    keys = list(Q.keys()) #you can actually do if v in Q:
    #check if v in keys
    return v in keys

while Q:
    #u = extract_min
    u = Q.popitem()[0]
    for v, weight in G[u]:
        if v_in_Q(Q, v) and weight < Q[v]:
            #v.pi = u
            pi[v] = u
            #v.key = w(u, v)
            Q[v] = weight
print("Final PI:", pi)
```
